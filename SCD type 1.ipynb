# SCD Type 1

In SCD type 1, the history of the changes will not be maintained in the dimension tanles. IN this type overwriting of existing data with new information which leads to storing only the current, active record without any history.

employee_stg is the staging table that get new or updated recoards on dialy basis, these recaords are further loaded to employee dimension table.

```python
%sql
create database scd;
create table employee_stg (employee_id int primary key, name string, salary int, dept string);
```

Below is the Employee dimension table which shows behaviour of SCD Type1

```python
%sql
create table employee (employee_id int primary key, name string, salary int, dept string);
```

```python
%sql
select * from employee;
select * from employee_stg;
```

Below code will update info of the existing employees, whos data has been updated in staging table.

```python
%sql
MERGE INTO employee AS dim
USING employee_stg AS stg
ON dim.employee_id = stg.employee_id
WHEN MATCHED THEN
  UPDATE SET dim.salary = stg.salary, dim.dept = stg.dept;
```
