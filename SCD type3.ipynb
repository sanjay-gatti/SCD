# SCD Type 3

SCD Type 3 stores both the current value and a limited number of previous values in the same row by adding new columns (e.g., current_dept, previous_dept). It preserves only partial history and is used when only the most recent change matters.

```python
%sql
create database scd;
```

```python
%sql
create table employee_stg (employee_id int primary key, name string, salary int, dept string);
create table employee (employee_id int primary key, name string, current_salary int, current_dept string,prev_salary int, prev_dept string );
```

Insert new records and update existing records to Employee Dimension

```python
%sql
MERGE INTO employee AS dim
USING employee_stg AS stg
ON dim.employee_id = stg.employee_id
WHEN MATCHED THEN
  UPDATE SET dim.prev_salary = dim.current_salary,
             dim.prev_dept = dim.current_dept,
             dim.current_salary = stg.salary,
             dim.current_dept = stg.dept
WHEN NOT MATCHED THEN
INSERT (employee_id, name, current_salary,current_dept, prev_salary, prev_dept) VALUES (stg.employee_id, stg.name, stg.salary, stg.dept, NULL, NULL)
```
